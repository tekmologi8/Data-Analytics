```python
import pandas as pd
import phonenumbers

# Load 20211129_test_data raw data
df = pd.read_csv("/content/20211129_test_data.csv", encoding= 'unicode_escape')
```

```python
df.head()
```

```python
# dropped all columns that are missing data values other than fullName column that we'll use later
df=df.drop(columns=['campaignId','addressLine2','emailAddress','asset3','asset4','asset5','customField4','customField5'])

df
```

```python
# Combines firstName and lastName and list them under the fullName column
df['fullName'] = df['firstName'].str.cat(df['lastName'],sep=" ")

df
```

```python
# Sorting by column "isPrimary". Record marked as ‘isPrimary’ = TRUE, should be addressed first. 
df.sort_values(by=['isPrimary'], ascending=False)
```

```python
#Combines households based on coordinatedMailingId, asset1, addressLine1 and isPrimary
gk = df.groupby(['coordinatedMailingId','asset1','addressLine1','isPrimary'])

gk.first()
```

```python
#drop ALL customerPersonId duplicate values and keep the last recorded value
df=df.drop_duplicates(subset = ['customerPersonId','asset1','asset2'], keep = 'last')

df
```

```python
from dataprep.clean import clean_phone
clean_phone(df, "phoneNumber")
```

```python
# Reformat the ‘phoneNumber’ column to add ‘-‘, (ex. 1-888-111-2222)
df['phoneNumber']

for phone_no in df['phoneNumber']: 
  contactphone = "%c-%c%c%c-%c%c%c-%c%c%c%c" % tuple(map(ord,list(str(phone_no)[:11])))

df['phoneNumber'] = contactphone
    
df
```

```python
# reformat the drop date to long format. (ex. November 29, 2021)
df['dropDate'] = pd.to_datetime(df['dropDate'], format='%m/%d/%Y')
df['dropDate'] = df['dropDate'].dt.strftime('%B %d, %Y')
df.head()
```
